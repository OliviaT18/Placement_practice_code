## Orbit of Halleys commet based on the Velocity Verlet algorithm

Based on a java example in 'Introduction to Computational physics'  by Tao Pang

learning:
* python basics
* algorithm implementation
* unit manipulation - rebasing
* the verlet algorithm

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# set some key parameters

# set time units based on 76 years - orbit time
# length unit set to 2.68x10^12 m  the semi major axis

v_min = 0.815795
r_max = 1.966843 
kappa = 39.5   # kappa = GM

n = 20000  # number of steps
h = 2.0/(n-1) # time step width
h2 = (h*h)/2.0 # from 1/2*a*t^2


# set up lists
t = [0]                         # time
x = [r_max]                     # x pos
y = [0]                         # y pos
r = [r_max]                     # r = sqrt(x^2 * y^2)
vx = [0]                        # x velocity component
vy = [v_min]                    # y velocity component
gx = [-1*kappa/(r_max*r_max)]   # x acceleration component
gy = [0]                        # y acceleration component

i = 0

while i < n:
    t.append(h*(i+1))                     # next time
    x.append(x[i]+h*vx[i]+h2*gx[i])       # next x pos
    y.append(y[i]+h*vy[i]+h2*gy[i])       # next y pos
    
    r2 = (x[i+1]*x[i+1])+(y[i+1]*y[i+1])  # next r squared
    r.append(np.sqrt(r2))                 # next r
    r3 = r2*r[i+1]                        # next r cubed
    
    gx.append(-1*kappa*x[i+1]/r3)
    gy.append(-1*kappa*y[i+1]/r3)
    vx.append(vx[i]+h*(gx[i+1]+gx[i])/2)  # next velocity x component
    vy.append(vy[i]+h*(gy[i+1]+gy[i])/2)  # next velocity y component
    
    i = i + 1
    
plt.plot(t, r)
plt.xlabel("number of orbits")
plt.ylabel(" radius, in semi major axis units")
   

```
