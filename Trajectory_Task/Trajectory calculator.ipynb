## Trajectory of a Ball


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as spc
```

```python
g = spc.g
u = 5
y0 = 0
theta = np.deg2rad(45)

def T(u, theta):
    t = (2 * u * np.sin(theta))/g
    t = np.linspace(0,2,30)
    return t
time = T(u, theta)
print ('time =', time)

def x_vel(u, theta):
    vx = u * np.cos(theta)
    return vx
x_velocity = x_vel(u, theta)
print ('x velocity =', x_velocity)
    
def y_vel(u, theta):
    vy = u * np.sin(theta)
    return vy
y_velocity = y_vel(u, theta)
print ('y velocity =', y_velocity)
    
def x_dis(x_velocity, time):
    sx = x_velocity * time
    return sx
x_displacement = x_dis(x_velocity, time)
print ('x displacement =', x_displacement) 

def y_dis(y_velocity, time, y0):
    sy = y_velocity * time - (0.5*g*(time**2)) + y0  
    return sy
y_displacement = y_dis(y_velocity, time, y0)
print ('y displacement =', y_displacement)

def R(u, theta):
    r = (u**2 * np.sin(2*theta))/ g
    return r
Range = R(u, theta)
print('range =', Range)

def height(u, theta):
    h = ((u**2) * (np.sin(theta))**2)/ (2*g)   
    return h
height = height(u, theta)
print ('max height =', height) 
       

plt.xlim(0,3.5)
plt.ylim(0,3.5)
plt.xlabel('x Displacement /m')
plt.ylabel('y Displacement /m')
plt.plot (x_displacement, y_displacement,'m-', label = theta)
plt.show()
```

## Neutron Displacement

```python

'neutron rest mass'
n = 1.675e-27

'converting eV to J'
E = [0.1, 0.01, 0.001]
Ej = [i*1.6e-19 for i in E]
print ('Energy =', Ej)

'calculating initial velocities'
nu = [np.sqrt((2*j)/n) for j in Ej]
print ('Initial velocity =', nu)

theta = np.deg2rad(0)
print (theta)
y0 = 0

nT = []
xv = []
yv = []
xs = []
ys = []


for U in nu:
    n_time = T(U, theta)
    n_time = np.linspace(0,2,200)
    nT.append(n_time)

    nx_velocity = x_vel(U, theta)
    xv.append(nx_velocity)
    print ('x velocity =', nx_velocity)

    ny_velocity = y_vel(U, theta)
    yv.append(ny_velocity)
    print ('y velocity =', ny_velocity)    

    nx_displacement = x_dis(nx_velocity, n_time)
    xs.append(nx_displacement)  

    ny_displacement = y_dis(ny_velocity, n_time, y0)
    ys.append(ny_displacement)

    plt.plot(nx_displacement, ny_displacement, label='%s m/s' % U)
    plt.xlim(0,96)
    plt.ylim(0,-0.3)
    plt.xlabel('x displacement /m')
    plt.ylabel('y displacement /m')
    plt.legend()


plt.show()

X = [96, 46, 11]

for x in X:
    iv = [u**2 for u in nu]
    for v in iv:
        y = y0 + x * np.tan(theta) + (g * (x**2))/(2 * v * ((np.cos(theta))**2))
    print('Distance from moderator =', x, 'm', '  ', 'Neutron displacement =', y, 'm')

    for U in nu:
        tn = x / U
        tt = 1
        dist = U * tt
    print('Distance travelled in', tt, 's =', dist, 'm')
    print('time =', tn, 's')
    

    

    
```
